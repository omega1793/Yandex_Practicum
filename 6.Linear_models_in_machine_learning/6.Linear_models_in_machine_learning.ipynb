# Проект: Линейные модели в машинном обучении

**Цель проекта** - разработать модель машинного обучения, которая поможет владельцу молочного хозяйства "Вольный луг" управлять рисками и принимать объективное решение о покупке коров у ассоциации пастбищ "ЭкоФерма" для расширения поголовья своего стада. 

В распоряжении 3 датафрйма:
* **ferma_main.csv** - содержит данные о стаде фермера на текущий момент:
   * Описание даных:
      * *`id`* — уникальный идентификатор коровы;
      * *`Удой, кг`* — масса молока, которую корова даёт в год (в килограммах);
      * *`ЭКЕ (Энергетическая кормовая единица)`* — показатель питательности корма коровы;
      * *`Сырой протеин, г`* — содержание сырого протеина в корме (в граммах);
      * *`СПО (Сахаро-протеиновое соотношение)`* — отношение сахара к протеину в корме коровы;
      * *`Порода`* — порода коровы;
      * *`Тип пастбища`* — ландшафт лугов, на которых паслась корова;
      * *`порода папы_быка`* — порода папы коровы;
      * *`Жирность,%`* — содержание жиров в молоке (в процентах);
      * *`Белок,%`* — содержание белков в молоке (в процентах);
      * *`Вкус молока`* — оценка вкуса по личным критериям фермера, бинарный признак (вкусно, не вкусно);
      * *`Возраст`* — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).
* **ferma_dad.csv** - хранит имя папы каждой коровы в стаде фермера^
   * Описание данных:
      * *`id* — уникальный идентификатор коровы;
      * *`Имя Папы* — имя папы коровы.
* **cow_buy.csv** - данные о коровах "ЭкоФермы":
   * Описание данных:
      * *`Порода`* — порода коровы;
      * *`Тип пастбища`* — ландшафт лугов, на которых паслась корова;
      * *`порода папы_быка`* — порода папы коровы;
      * *`Имя_папы`* — имя папы коровы;
      * *`Текущая_жирность,%`* — содержание жиров в молоке (в процентах);
      * *`Текущий_уровень_белок,%`* — содержание белков в молоке (в процентах);
      * *`Возраст`* — возраст коровы, бинарный признак (менее_2_лет, более_2_лет). 
      
      
Данные в файлах **ferma_main.csv** и **cow_buy.csv** похожи, но у датафрейма **cow_buy.csv** несколько особенностей:
* Содержание *белков* и *жиров* в молоке указано на момент продажи — когда "ЭкоФерма" давала коровам свой корм;
* Параметры корма *ЭКЕ (Энергетическая кормовая единица)*, *Сырой протеин, г* и *СПО (Сахаро-протеиновое соотношение)* отсутствуют. Технологи заказчика пересмотрели подход к кормлению: для новых коров планируется увеличить значения каждого из этих параметров на 5%;

```python
# Объявим функцию предобработки данных и обучение модели
def create_model_pipeline(numeric_features, categorical_features, model):
    numeric_transformer = Pipeline(steps=[
        ('scaler', StandardScaler())
    ])
    
    categorical_transformer = Pipeline(steps=[
        ('onehot', OneHotEncoder(drop='first',sparse=False, handle_unknown='error'))
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ])

    model_pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('pipeline', model)
    ])
    return model_pipeline
```

### Сравнение трёх моделей

```python
# Объявим функцию, считающую метрики и доверительный интервал модели
def metrics(model, y_pred, y_test):
    r_squared = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    interval = round(np.quantile(y_pred, 0.025)) , round(np.quantile(y_pred, 0.975))
    print(f"Модель: {model}")
    print(f"Метрика R²: {r_squared}")
    print(f"Метрика MSE: {mse}")
    print(f"Метрика MAE: {mae}")
    print(f"Метрика RMSE: {rmse}")
    print(f"Доверительный интервал модели: {interval}")
    return 
```

```python
# Выведем все метрики на экран
metrics('model_1', y_pred_1, y_test_1)
print()
metrics('model_2', y_pred_2, y_test_2)
print()
metrics('model_3', y_pred_3, y_test_3)
```

По показателям всех метрик выходит, что лучшая модель получилась **model_3**. Её доверительный интервал начинается от **5295кг** и заканчивается **6749кг**. Эту модель и будем использовать для прогнозирования удоя.

```python
# Создадим датафрейм с вероятностями
y_proba_log = model_log.predict_proba(X_test_log)[:,1]

data = pd.DataFrame(zip(y_test_log, y_proba_log),
             columns = ['y_valid', 'y_proba'])

data.head()
```

```python
# Создадим список порогов
thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 99,endpoint=False)]
```

```python
# Создадим переменную, которая будет хранить список новых колонок
columns = []

# Добавим столбцы с новыми предсказаниями в таблицу
for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_'+str(i)] = data['y_proba'].apply(lambda x:1 if x>=i else 0)

# Выведем 5 случайных строк
data.sample(5)
```

```python
# Построим график значений Precision  в зависимостии от порога

# Список порогов
thresholds = np.linspace(0, 1, 100)  

# Список значений Precision
precisions = []

# Precision для каждого порога
for threshold in thresholds:
    y_pred_log = (y_proba_log > threshold).astype(int)
    precision = precision_score(y_test_log, y_pred_log,zero_division=0)
    precisions.append(precision)

# график зависимости Precision от порога
plt.plot(thresholds, precisions)
plt.xlabel('Порог')
plt.ylabel('Precision')
plt.title('График зависимости Precision от порога')
plt.show()
```

По графику видно, что значение **precision** равно **1** при **пороге** в **0,7**. Это значение порога и использует при прогнозе вкуса молока.
