```python
# –û–±—ä—è–≤–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è –≤—Å–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ–± –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
def df_review(df):
    print('–ö—Ä–∞—Ç–∫–æ–µ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞')
    display(df.head())
    print('='*100)    
    print('–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ')
    df.info()
    print('='*100)   
    print('–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞:', df.shape)
    print('='*100)  
    print('–û—Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑–∞—Ü–∏—è —á–∏—Å–ª–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞')
    display(df.describe().T)
    print('='*100)
    print('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –≤—Å–µ—Ö —á–∏—Å–ª–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö')
    df.hist(figsize=(10, 5), bins=24);
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    plt.show()
    print('='*100)
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π—Ç–µ:', df.duplicated().sum())
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π—Ç–µ:', '\n', df.isna().sum())
    
# –û–±—ä—è–≤–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–µ–∫–æ—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
def decorate_plot(plt, title, xlabel = ' –î–∞—Ç–∞', ylabel = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤'):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)

# –û–±—ä—è–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞ –Ω–∞ —Ç—Ä–µ–Ω–¥ –∏ —Å–µ–∑–æ–Ω–Ω—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É
def drow_decomposed_data(data, suptitle):
    decomposed = seasonal_decompose(data)
    fig = plt.figure(figsize=(20, 20))
    # –ì—Ä–∞—Ñ–∏–∫ —Ç—Ä–µ–Ω–¥–∞
    plt.subplot(311)
    decomposed.trend.plot(ax=plt.gca())
    decorate_plot(plt, '–¢—Ä–µ–Ω–¥')
    # –ì—Ä–∞—Ñ–∏–∫ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏
    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca()) 
    decorate_plot(plt, '–°–µ–∑–æ–Ω–Ω–æ—Å—Ç—å')
    # –ì—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤
    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca()) 
    decorate_plot(plt, '–û—Å—Ç–∞—Ç–æ–∫')
    plt.tight_layout()
    fig.suptitle(suptitle)  
    plt.show()

# –û–±—ä—è–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–µ—Ä–µ–¥ –æ–±—É—á–µ–Ω–∏–µ–º –º–æ–¥–µ–ª–∏
def add_features(data, lags_count):
    data = df.copy()
    data['month'] = df.index.month
    data['day'] = df.index.day
    data['dayofweek'] = df.index.dayofweek
    data['hour'] = df.index.hour
    for i in range(1, lags_count + 1):
        data['lag_{}'.format(i)] = data['num_orders'].shift(i)  
    data['rolling_mean'] = data['num_orders'].shift().rolling(1).mean()    
    # –£–¥–∞–ª—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –ª–∞–≥–∞—Ö
    data = data.dropna()
    data = data.reset_index(drop=True)    
    return data

# –û–±—ä—è–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏
def fit_model(estimator, param_grid, X_train, y_train, X_test, y_test):
    model = GridSearchCV(estimator=estimator,
                         param_grid=param_grid,
                         n_jobs=-1,
                         cv=TimeSeriesSplit(n_splits=5),
                         scoring='neg_root_mean_squared_error'
                         )
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time
    best_rmse = abs(round(model.best_score_, 1))
    print(f'–õ—É—á—à–∞—è RMSE: {best_rmse}')
    print(f'–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏: {model.best_params_}')
    best_model = estimator.set_params(**model.best_params_)
    start_time = time.time()
    best_model.fit(X_train, y_train)
    predict_time = time.time() - start_time
    predict_train = best_model.predict(X_train)
    train_rmse = mean_squared_error(y_train, predict_train, squared=False)
    return best_model, best_rmse, train_time, predict_time, train_rmse
```
